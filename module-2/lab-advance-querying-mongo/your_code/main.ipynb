# Advanced Querying Mongo

Importing libraries and setting up connection

```python
from pymongo import MongoClient
conn = MongoClient("localhost:27017")
```

```python
db = conn.get_database("ironhack")
db.list_collection_names()
```

```python
collection = db.get_collection("companies")
```

### 1. All the companies whose name match 'Babelgum'. Retrieve only their `name` field.

```python
# Print also the id to see better that there's just one, even though the sintaxys would be {"name": 1, "_id": 0} 
list(collection.find({"name": "Babelgum"}, {"name": 1}))
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
# Print just the name and _id because they occupy so much space even being limited to 20.
collection.find({"number_of_employees": {"$gt": 5000}}, {"name": 1, "_id": 0}).limit(20).sort("number_of_employees", -1)
```

```python
list(_)
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fields.

```python
list(collection.find({"founded_year": {"$gte": 2000, "$lte": 2005}}, {"name": 1, "founded_year": 1, "_id": 0}))
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
list(collection.find({"ipo.valuation_amount": {"$gt": 100000000}, "founded_year": {"$lt": 2010}}, {"name":1, "ipo":1, "_id": 0}))
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
# Print just the name and _id because they occupy so much space even being limited to 10.
collection.find({"number_of_employees": {"$lt": 1000}, "founded_year": {"$lt": 2005}}, {"name": 1}).sort("number_of_employees", -1).limit(10)
```

```python
list(_)
```
